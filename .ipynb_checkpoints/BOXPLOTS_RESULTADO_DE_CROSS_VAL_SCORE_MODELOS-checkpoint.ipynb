En este notebook creo la gráfica de boxplot con los resultados de las distintas iteraciones de cross validation para los distintos modelos a comparar en la selección de modelo de este trabajo. Importo las librerías:

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from matplotlib.pyplot import axes
```

Cargo los datos:

```python
df_cv = pd.read_csv('resultados_cv.csv') # Dataset donde he guardado los negative log loss de las diferetentes 
                                         # iteraciones del cross_val_score para cada uno de los modelos a comparar.
```

```python
df_cv # Contenido del dataset.
```

Creo una lista de modelos para usarla posteriormente en el gráfico generado con matplotlib:

```python
modelos = list(df_cv['modelos'].unique()) 
```

```python
modelos
```

Elimino KNN, ya que en la selección de modelos, KNN dio el peor resultado, y muy alejado del resto, lo que dificulta la interpretabilidad y claridad de la visualización si se lo incluye:

```python
modelos.pop(3)
```

 Gráfico con cinco boxplots, uno por cada modelo:

```python
plt.figure(figsize=(12,8))

datos = [] # Creo una lista.

for i in modelos: # Por cada modelo
    
    model_plot = df_cv[df_cv['modelos']== i] # Filtro en el dataset los registros del modelo
    datos.append(model_plot['neg_log_loss']) # Y añado a la lista, el array de los distintos valores 
                                             # de negative log loss del modelo.
    
axes().set_xticklabels(modelos, rotation=90) 

plt.boxplot(datos, showmeans=True) # Represento los datos de negative log loss de cada modelo en forma de boxplot.

plt.ylim(-0.7, -0.6)


plt.xlabel('Modelos', fontsize=14)
plt.ylabel('neg.log loss', fontsize=14)
plt.title('Boxplots sobre los resultados de cross validation en los modelos', fontsize=16)
plt.show()
```
